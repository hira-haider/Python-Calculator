```python
import tkinter as tk
from tkinter import messagebox

def calculate():
    try:
        a = int(entry1.get())
        b = int(entry2.get())
        operation = operator_var.get()

        result = None
        if operation == "+":
            result = a + b
        elif operation == "-":
            result = a - b
        elif operation == "*":
            result = a * b
        elif operation == "/":
            if b == 0:
                messagebox.showerror("Error", "Division by zero!")
                return
            else:
                result = a / b
        elif operation == "s":
            result = (a ** 0.5, b ** 0.5)
        elif operation == "^":
            result = a ** b
        else:
            messagebox.showerror("Error", "Invalid operation!")
            return

        if operation == "s":
            result_label.config(text=f"√{a} = {result[0]}\n√{b} = {result[1]}")
        else:
            result_label.config(text=f"Result: {result}")
    except ValueError:
        messagebox.showerror("Error", "Invalid input!")

# Create the main window
window = tk.Tk()
window.title("Calculator")

# Create input fields
entry1_label = tk.Label(window, text="Enter 1st number:")
entry1_label.grid(row=0, column=0)
entry1 = tk.Entry(window)
entry1.grid(row=0, column=1)

entry2_label = tk.Label(window, text="Enter 2nd number:")
entry2_label.grid(row=1, column=0)
entry2 = tk.Entry(window)
entry2.grid(row=1, column=1)

# Create operator selection
operator_var = tk.StringVar(window)
operator_var.set("+")
operator_label = tk.Label(window, text="Select operation:")
operator_label.grid(row=2, column=0)
operator_menu = tk.OptionMenu(window, operator_var, "+", "-", "*", "/", "s", "^")
operator_menu.grid(row=2, column=1)

# Create calculate button
calculate_button = tk.Button(window, text="Calculate", command=calculate)
calculate_button.grid(row=3, columnspan=2)

# Create result label
result_label = tk.Label(window, text="")
result_label.grid(row=4, columnspan=2)

# Run the GUI loop
window.mainloop()
```
